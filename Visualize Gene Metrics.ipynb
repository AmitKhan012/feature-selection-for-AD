```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("dataset/Gene_Metrics_Info.csv")
col = df["Gene Symbol"]
# df.drop("Gene Symbol", axis =1, inplace = True)
df2 = df.drop("Gene Symbol", axis =1)
df2 =df2.T.rename(columns = col)
```

```python
df2
```

```python
def plot_graph(column_stats):
    column_stats = column_stats.T

    # Create a stacked bar chart
    fig, ax = plt.subplots(figsize=(12, 6))

    column_stats.plot(kind='bar', stacked=True, ax=ax)
    plt.xticks(rotation=90)

    plt.show()
```

```python
for i in range(len(df2)):
  plot_graph(df2.iloc[i:i+1])
```
